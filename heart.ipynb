```python
features=['sex','trestbps','chol','fbs','restecg','thalach','exang','oldpeak','slope','ca','thal']
plt.figure(figsize=(15,12)) # Adjusted figsize for better visualization with more subplots
for i, feature in enumerate(features,1):
  plt.subplot(3,4,i) # Changed to 3 rows, 4 columns to fit all 11 features
  sns.boxplot(x='target',y=feature,data=df)
  plt.title(f'boxplot of {feature} vs Target') # Improved title formatting
plt.tight_layout()
plt.show()
```
